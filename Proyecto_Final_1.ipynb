Importar y preparar datos

```python
Archivo="Clean_Top_1000_Youtube_df - youtubers_df.csv"
dataset= pd.read_csv(Archivo)
```

```python
print(dataset.shape)
dataset.head()
```

```python
#Variables categoricas y númericas
dataset.info()
```

```python
#Eliminar filas con datos faltantes
dataset.dropna(inplace=True)
dataset.info()
```

```python
#Conteo de los niveles en las diferentes columnas categóricas
cols_cat = ['Username', 'Categories', 'Suscribers', 'Country', 'Visits', 'Likes', 'Comments', 'Links']
for col in cols_cat:
  print(f'Columna {col}: {dataset[col].nunique()} subniveles')
```

```python
dataset.describe()
```

```python
#Eliminar filas repetidas
print(f'Tamaño del set antes de eliminar las filas repetidas: {dataset.shape}')
dataset.drop_duplicates (inplace=True)
print(f'Tamaño del set después de eliminar las filas repetidas: {dataset.shape}')
```

```python
check_dup = dataset.duplicated().any()
print("¿Existen valores duplicados en los datos?",check_dup)
if check_dup:
  dataser = dataset.drop_duplicates()
  print("Los valores duplicados fueron elimininados")
else:
  print("No hay valores duplicados en los datos")
```

```python
dataset.drop(['Links'], axis=1, inplace=True)
```
